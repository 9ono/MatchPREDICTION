```python
driver = webdriver.Chrome('../chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(3)
all_match = []
for i in matches:
    for j in i:
        driver.get('https://www.whoscored.com' + j)
        try:
            driver.find_element_by_xpath('//*[@id="qcCmpButtons"]/button[2]').click()
        except:
            pass


        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        team_h = soup.select('#match-centre-header > div:nth-child(1) > div.team-info > a')
        team_a = soup.select('#match-centre-header > div:nth-child(3) > div.team-info > a')
        score = soup.select('#match-header > table > tbody > tr:nth-child(1) > td.result')
        #driver.find_element_by_xpath('/html/body/div[6]/div[2]/div[5]/div[2]/div[2]/ul/li[3]/div[2]/span').click()
        #time.sleep(3)
        m = soup.find_all('span')
    
        for i in team_h[0].stripped_strings:
            home = i
        for i in team_a[0].stripped_strings:
            away = i
    
        for i in score[0].stripped_strings:
            sco = i
        sco = sco.replace(' ','')
        sco = sco.split(':')
        print(sco)
        #print(home, away)

        match_stat = []
        tmp = []
        for i in m:
            if 'class' in i.attrs:
                a = str(i.attrs)
                if 'home' in a:
                    b = a
                    b = b[-6:-1].replace('\'','')
                    b = b.replace(' ','')
                    b = b.replace(':','')
                    #print(b)
                    tmp.append(b)
                if 'away' in a:
                    b = a
                    b = b[-6:-1].replace('\'','')
                    b = b.replace(' ','')
                    b = b.replace(':','')
                    #print(b)
                    tmp.append(b)
                    match_stat.append(tmp)
                    tmp = []

        one_match = [home, away, sco[0], sco[1]]
        for i in range(40):
            if i != 1 and i != 7 and i != 10 and i != 15 and i != 20 and i != 25 and i != 32 and i != 35:
                one_match.extend(match_stat[i])
        print(one_match)
        all_match.append(one_match)
```
